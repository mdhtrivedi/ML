Using Label Encoding, convert categories to numeric

```python
dfx["Marital status"] = dfx["Marital status"].cat.codes
dfx["Application mode"] = dfx["Application mode"].cat.codes
dfx["Course"] = dfx["Course"].cat.codes
dfx["MDaytime/evening attendance"] = dfx["MDaytime/evening attendance"].cat.codes
dfx["Previous qualification"] = dfx["Previous qualification"].cat.codes
dfx["Nacionality"] = dfx["Nacionality"].cat.codes
dfx["Mother's qualification"] = dfx["Mother's qualification"].cat.codes
dfx["Father's qualification"] = dfx["Father's qualification"].cat.codes
dfx["Mother's occupation"] = dfx["Mother's occupation"].cat.codes
dfx["Father's occupation"] = dfx["Father's occupation"].cat.codes
dfx["Displaced"] = dfx["Displaced"].cat.codes
dfx["Educational special needs"] = dfx["Educational special needs"].cat.codes
dfx["Debtor"] = dfx["Debtor"].cat.codes
dfx["Tuition fees up to date"] = dfx["Tuition fees up to date"].cat.codes
dfx["Gender"] = dfx["Gender"].cat.codes
dfx["Scholarship holder"] = dfx["Scholarship holder"].cat.codes
dfx["International"] = dfx["International"].cat.codes
```

Normalize input dataset

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler_dfx = pd.DataFrame(scaler.fit_transform(dfx),columns = dfx.columns)
scaler_dfx.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_train,input_test,output_train,output_test = train_test_split(scaler_dfx,dfy,test_size = 0.2, random_state = 1)
```

```python
input_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier()
KNN
```

```python
Knn = KNN.fit(input_train,output_train)
Knn
```

```python
pred_test = Knn.predict(input_test)
pred_test
```

```python
output_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score_knn = round(accuracy_score(output_test,pred_test)*100,0)
accuracy_score_knn
```

Linear Regression

```python
dict = {"Unemployment": studentdf['Unemployment rate'],
        "GDP": studentdf['GDP']}
```

```python
df = pd.DataFrame(dict)
```

```python
df.head()
```

```python
x = df.drop("Unemployment", axis=1)
y = df["Unemployment"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 3)
```

```python
x_train.shape[0]
```

```python
x_test.shape[0]
```

```python
x_train.head()
```

```python
from  sklearn.linear_model import LinearRegression
```

```python
Model = regressor.fit(x,y)
```

```python
Model
```

```python
Model.coef_
```

```python
Model.intercept_
```

```python
plt.scatter(x_train["GDP"], y_train, color='red') #Plotting observation line
plt.plot(x_train["GDP"], Model.predict(x_train), color='blue') #plotting regression line
plt.title("Unemployment vs GDP (Training Set)")
plt.xlabel("GDP")
plt.ylabel("Unemployment")
plt.show()
```
